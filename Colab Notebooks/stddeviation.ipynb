```python
import csv
from math import sqrt

# Membaca data dari file csv
data = []
with open('irisv2.csv') as file:
  reader = csv.reader(file)
  next(reader) # Skip header
  for row in reader:
    data.append([float(row[0])])

# Menghitung rata rata dari setiap kolom
mean = sum(col[0] for col in data) / len(data)

# Menghitung jarak antara setiap nilai dari kolom dan rata rata kolom
differences = [(col[0] - mean) ** 2 for col in data] 

# Menghitung variasi dari setiap kolom
variance = sum(differences) / len(data)

# Menghitung standar deviasi
stddev = sqrt(variance)
print("Standar deviasi dari sepal lenght: ", stddev)
```
