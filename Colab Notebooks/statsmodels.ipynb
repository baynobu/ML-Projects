```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# load dataset
df = pd.read_csv('/content/train.csv')

df['x1'] = 0 #buat kolom x dengan nilai 0
x1 = df['x1'] #buat variabel X utk menyimpan kolom x
for i in range(0, len(df)):
  x1[i] = i+1

# define independent and dependent variables
X = df[['x', 'x1']]
y = df['y']

# add constant to independent variables
X = sm.add_constant(X)

# create linear regression model
model = sm.OLS(y, X)

# fit the model
results = model.fit()

# print summary of the model
print(results.summary())
```

```python
# load dataset
df = pd.read_csv('train.csv', index_col='y', parse_dates=True)

# create ARIMA model
model = sm.tsa.ARIMA(df, order=(1,1,1))

# fit the model
results = model.fit()

# print summary of the model
print(results.summary())
```

```python
# load dataset
df = pd.read_csv('train.csv')

# define independent and dependent variables
X = df[['x1', 'x2']]
y = df['y']

# create multivariate regression model
model = sm.OLS(y, X)

# fit the model
results = model.fit()

# print summary of the model
print(results.summary())
```
